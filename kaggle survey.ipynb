```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv('kaggler.csv')
```

```python
df.head()
```

```python
df = df.drop(0)
```

```python
df = df.rename(columns = {"Time from Start to Finish (seconds)": "duration","Q1":"age","Q2":"gender","Q3":"country","Q4":"degree","Q5":"profession","Q6":"exp"}, inplace = False)
```

```python
df.head()
```

```python
df['duration'].isnull().sum()
```

```python
df['duration']=df['duration'].astype(int)
```

```python
df['duration']=df['duration']/60
```

```python
sns.distplot(df['duration'],color='yellow')
plt.title('duration')
plt.show()
```

```python
df['duration'].describe()
```

```python
df.loc[df['duration']<600]
```

```python
plt.scatter(df['duration'],df['age'],color='red')
plt.title('duration/age')
plt.show()
```
